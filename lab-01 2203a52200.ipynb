```python
!pip install matplotlib
import matplotlib.pyplot as plt # Import the matplotlib library and assign it to the alias 'plt'

age_filter = (c['age'] >= 20) | (c['age'] <= 40)
# next lets filter for smoker
smoker_filter = (c['smoker'] == 'yes') | (c['smoker'] == 'no')
# finally lets combine both filters
age_filtered = c[age_filter & smoker_filter]
print(age_filtered)
# 4. Plot the filtered data
plt.figure(figsize=(10, 6)) # Now plt is recognized and the figure can be created

# Scatter plot: Plot age vs expenses, differentiating smokers and non-smokers
smokers = age_filtered[age_filtered['smoker'] == 'yes']
non_smokers = age_filtered[age_filtered['smoker'] == 'no']

# Plot smokers in red
plt.scatter(smokers['age'], smokers['expenses'], c='red', label='Smokers', marker='o')

# Plot non-smokers in blue
plt.scatter(non_smokers['age'], non_smokers['expenses'], c='blue', label='Non-Smokers', marker='x')

# Add labels and title
plt.title('Age vs Expenses for Smokers and Non-Smokers (Age 20-40)', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Expenses', fontsize=12)

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file into the DataFrame
df = pd.read_csv('/content/insurance.csv')  # This line reads the csv into a dataframe

# Calculate average expenses for smokers and non-smokers
avg_expenses = df.groupby("smoker")["expenses"].mean()

# Plot the data
plt.figure(figsize=(8, 6))
avg_expenses.plot(kind="bar", color=["blue", "green"], alpha=0.7)
plt.xlabel("Smoking Status")
plt.ylabel("Average Expenses")
plt.title("Average Medical Expenses of Smokers vs Non-Smokers")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/insurance.csv')

# Count the number of smokers grouped by gender
smokers_gender = df[df["smoker"] == "yes"].groupby("sex")["smoker"].count()

# Plot the data
plt.figure(figsize=(8, 6))
smokers_gender.plot(kind="bar", color=["purple", "orange"], alpha=0.7)
plt.xlabel("Gender")
plt.ylabel("Number of Smokers")
plt.title("Number of Smokers by Gender")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
